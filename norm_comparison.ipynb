```python
import os
import numpy as np
import matplotlib.pyplot as plt

pic_dir = "/home/mehak-gupta/mehak/project/report/pics/final/"
normalized_dir = "/home/mehak-gupta/mehak/project/UDF_HUDF_sources_FULL /bg_sources/stline_cont/normalized_output1/"

# Specify the exact four files you want to plot
selected_files = [
    "norm_29.909_1051.txt",
    "norm_23.645_50.txt",
    "norm_26.709_1016.txt",
    "norm_29.815_21.txt"
]

# Sanity check (optional)
for f in selected_files:
    if not os.path.isfile(os.path.join(normalized_dir, f)):
        print(f"File not found: {f}")

# Set up subplot grid: 2 rows × 2 columns
fig, axes = plt.subplots(2, 2, figsize=(10, 5))
axes = axes.flatten()

for idx, (ax, filename) in enumerate(zip(axes, selected_files)):
    file_path = os.path.join(normalized_dir, filename)

    # Load the normalized spectrum
    data = np.loadtxt(file_path)
    shifted_wave = data[:, 0]
    norm_flux    = data[:, 1]
    norm_error   = data[:, 2]

    # Plot
    ax.plot(shifted_wave, norm_flux, color='navy')
    ax.fill_between(shifted_wave, norm_flux - norm_error, norm_flux + norm_error,
                    color='lightblue', alpha=0.5)
    ax.set_xlim(2700, 2900)
    ax.set_ylim(-2, 4)
    ax.set_title(filename, fontsize=10)
    ax.grid(True, linestyle='--', alpha=0.4)
    ax.set_xlabel("Wavelength (Å)")
    ax.set_ylabel("Normalized Flux")

plt.tight_layout()
filepath = os.path.join(pic_dir, "norm_st_ran4.png")
plt.savefig(filepath, dpi=300, bbox_inches='tight')
plt.show()
```
