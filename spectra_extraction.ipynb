```python
import os
from astropy.io import fits
import numpy as np

fits_directory = '/home/mehak-gupta/mehak/project/UDF_HUDF_sources_FULL /sources_fits/'
ascii_output_directory = '/home/mehak-gupta/mehak/project/UDF_HUDF_sources_FULL /sources_ascii_MgII'

# Ensuring the output directory exists
os.makedirs(ascii_output_directory, exist_ok=True)

# List to store all source IDs
source_ids = []

# Function to extract and save data to ASCII
def fits_to_ascii(fits_file, output_directory):
    with fits.open(fits_file) as hdul:
        # Extracting the source ID
        source_id = hdul[0].header.get('ID', 'UnknownID')  # Default to 'UnknownID' if ID is missing

        # Add source_id to the list
        source_ids.append(source_id)

        # Checking if HDU[1] is named "Z"
        if hdul[1].name == 'Z':
            flux = hdul[2].data  # HDU[2] as flux
            var = hdul[3].data   # HDU[3] as variance
            pl_fit = hdul[8].data  # HDU[8] as pl_fit
            #pl_fitp = hdul[9].data  # HDU[9] as pl_fitp
        else:
            flux = hdul[1].data  # HDU[1] as flux
            var = hdul[2].data   # HDU[2] as variance
            pl_fit = hdul[7].data  # HDU[7] as pl_fit
            #pl_fitp = hdul[8].data  # HDU[8] as pl_fitp
            
        # Extracting header information for wavelength array from HDU[2]
        header = hdul[2].header
        min_w = header['CRVAL1']  # Starting wavelength
        step_w = header['CDELT1']  # Increment per pixel
        refpx = header['CRPIX1']  # Reference pixel (1-based)

        # Defining the wavelength array
        n_pixels = len(flux)  # Number of pixels in the flux array
        wavelength = min_w + (np.arange(n_pixels) + 1 - refpx) * step_w
        #print(wavelength.shape, flux.shape, var.shape, pl_fit.shape,source_id)
        # Combining wavelength, flux, variance, and pl_fit into a single array
        data = np.column_stack((wavelength, flux, var, pl_fit))
        
        # Creating the ASCII file name using the Source ID
        ascii_file = os.path.join(output_directory, f"{source_id}.txt")
        
        # Saving the data to an ASCII file
        np.savetxt(ascii_file, data, header="Wavelength     Flux       Variance      Pl_Fit    ", fmt="%.6e")

# Looping through all FITS files in the directory
for fits_file in os.listdir(fits_directory):
    if fits_file.endswith('.fits'):
        fits_path = os.path.join(fits_directory, fits_file)
        fits_to_ascii(fits_path, ascii_output_directory)

# Sorting source IDs in ascending order
source_ids.sort()

# Saving all sorted source IDs to a separate text file
with open(os.path.join(ascii_output_directory, 'source_ids.txt'), 'w') as file:
    for source_id in source_ids:
        file.write(f"{source_id}\n")

print("Conversion complete! Source IDs saved to 'source_ids.txt' in ascending order.")
```

```python
import os
from astropy.io import fits
import numpy as np

# File paths
bg_id_file = '/home/mehak-gupta/mehak/project/UDF_HUDF_Catalogs_Astropy_FULL/bg_id.txt'
source_id_file = '/home/mehak-gupta/mehak/project/UDF_HUDF_sources_FULL /sources_ascii/source_ids.txt'

# Reading the IDs from bg_id.txt
with open(bg_id_file, 'r') as file:
    bg_ids = set(line.strip() for line in file.readlines())

# Reading the IDs from source_ids.txt
with open(source_id_file, 'r') as file:
    source_ids = set(line.strip() for line in file.readlines())

# Finding common IDs (present in both bg_id.txt and source_ids.txt)
common_ids = bg_ids.intersection(source_ids)

# Finding IDs that are in bg_id.txt but not in source_ids.txt
bg_not_in_source = bg_ids.difference(source_ids)

# Converting IDs to integers for proper numerical sorting
common_ids = sorted([int(id) for id in common_ids])
bg_not_in_source = sorted([int(id) for id in bg_not_in_source])

# Saving the sorted common IDs to a file
with open('/home/mehak/mehak/project/UDF_HUDF_sources_FULL /sources_ascii/common_ids.txt', 'w') as file:
    for id in common_ids:
        file.write(f"{id}\n")

# Saving the sorted unique IDs to a file
with open('/home/mehak/mehak/project/UDF_HUDF_sources_FULL /sources_ascii/bg_not_in_source.txt', 'w') as file:
    for id in bg_not_in_source:
        file.write(f"{id}\n")

print("IDs saved to files in proper ascending numerical order.")
```

```python
import shutil
import os

# Define file paths
txt_file_path = '/home/mehak-gupta/mehak/project/UDF_HUDF_Catalogs_Astropy_FULL/bg_id_FeII_2374_2382.txt'  # Path to the txt file containing numbers
source_folder = '/home/mehak-gupta/mehak/project/UDF_HUDF_sources_FULL /sources_ascii/'  # Path to the folder containing the ASCII files
destination_folder = '/home/mehak-gupta/mehak/project/UDF_HUDF_sources_FULL /FeII_2374_2382/'  # Path to the folder where files should be copied

# Ensure destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Read the numbers from the txt file
with open(txt_file_path, 'r') as f:
    numbers = f.readlines()

# Loop through each number in the file
for number in numbers:
    # Clean up any extra whitespace (like newline characters)
    number = number.strip()

    # Construct the corresponding ASCII file name
    ascii_file_name = f"{number}.txt"  # Adjust the extension if needed

    # Construct the full path to the source file
    source_file = os.path.join(source_folder, ascii_file_name)

    # Check if the source file exists
    if os.path.exists(source_file):
        # Construct the full path for the destination file
        destination_file = os.path.join(destination_folder, ascii_file_name)

        # Copy the file to the destination folder
        shutil.copy(source_file, destination_file)
        #print(f"Copied {ascii_file_name} to {destination_folder}")
    else:
        print(f"File {ascii_file_name} not found in the source folder.")
```
